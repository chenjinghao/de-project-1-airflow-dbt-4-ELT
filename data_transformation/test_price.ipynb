```python
import pandas as pd
import json
```

```python
with open('0_PCLA_stocks_price.json', 'r') as f:
    data = json.load(f)

```

```python
print(data['Meta Data'])
```

```python
stock_meta = {key: data['Meta Data'][key] for key in ['2. Symbol', '3. Last Refreshed']}
stock_meta = pd.DataFrame([stock_meta])
print(stock_meta)
```

```python
stock_meta.rename(columns={'2. Symbol': 'symbol', '3. Last Refreshed': 'last_refreshed'}, inplace=True)
print(stock_meta)
```

```python
# convert symbol to primary key and last refreshed to datetime
stock_meta['symbol'] = stock_meta['symbol'].astype('string')
stock_meta['last_refreshed'] = pd.to_datetime(stock_meta['last_refreshed'])
print(stock_meta)
```

```python
stock_price = pd.DataFrame.from_dict(data['Time Series (Daily)'], orient='index')
print(stock_price.head())
```

```python
# change column names
stock_price.rename(columns={'1. open': 'open',
                            '2. high': 'high',
                            '3. low': 'low',
                            '4. close': 'close',
                            '5. volume': 'volume'}, inplace=True)
print(stock_price.head())
```

```python
# convert to appropriate data types
stock_price.index = pd.to_datetime(stock_price.index)
stock_price = stock_price.astype({'open': 'float', 'high': 'float', 'low': 'float', 'close': 'float', 'volume': 'int'})
print(stock_price.dtypes)
```
